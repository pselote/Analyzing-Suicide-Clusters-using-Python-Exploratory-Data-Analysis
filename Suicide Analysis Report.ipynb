#  DATA Analysis

# 1. Query to extraxt data on basis of gender

### * Total number of sucide cases by gender

```python
df.groupby("sex")["suicides_no"].sum()
```

### * Average suicide rates by gender

```python
df.groupby("sex")["suicides/100k pop"].mean()
```

### * Suicide Numbers by Gender

```python
plt.figure(figsize=(8,5))
sns.barplot(x = "sex", 
            y = "suicides_no", 
            data = df)
plt.xlabel('Gender')
plt.ylabel('Number of Suicide')
plt.title('Suicide Numbers by Gender')
plt.show()
```

* Clearly we can see that sucide rate of male are more than female.

# 2. Visualize the ranking of the countries by the no. of suicides

### * Average numbers of suicide by country
### * The countries with the highest numbers of suicide are below

```python
df.groupby("country", as_index = False)["suicides_no"].mean().sort_values(by="suicides_no",ascending = False).head()
```

### Bar Graph for Suicide Numbers by Countries

```python

countrySN = df.groupby("country", as_index = False)["suicides_no"].mean().sort_values(by="suicides_no",ascending = False)
plt.figure(figsize=(6,15))
sns.barplot(x = "suicides_no", 
            y = "country", 
            data = countrySN)
plt.xlabel('Number of Suicide')
plt.ylabel('Country')
plt.title('Suicide Numbers by Countries')
plt.show()
```

* Clearly we can see that Russian Federation has heighest number of sucide.

# 3. Which year highest number of sucide

### * Average numbers of suicide by year

```python
df.groupby("year", as_index = False)["suicides_no"].mean().sort_values(by="suicides_no",ascending = False)
```

### * Pointplot for Average Suicide Numbers by Years

```python
yearMeanSN = df.groupby("year", as_index = False)["suicides_no"].mean()
plt.figure(figsize=(15,6))
sns.pointplot(x = "year", 
              y = "suicides_no", 
              data = yearMeanSN)
plt.xlabel('Year')
plt.ylabel('Average Number of Suicide')
plt.title('Average Suicide Numbers by Years')
plt.show()
```

* We can clearly see that year 1994 has heighest number of sucides rate.

# 4. Which Age group has heighest number of suicides rate.

### * Average suicide rates by age

```python
df.groupby("age", as_index = False)["suicides/100k pop"].mean().sort_values(by="suicides/100k pop",ascending = False)
```

### * Barplot for Suicide Numbers by Age Groups

```python
ageDataSN = df.groupby("age", as_index = False)["suicides_no"].mean().sort_values(by="suicides_no",ascending = False)
plt.figure(figsize=(8,5))
sns.barplot(x = "age", 
            y = "suicides_no", 
            data = ageDataSN)
plt.xlabel('Age Group')
plt.ylabel('Number of Suicide')
plt.title('Suicide Numbers by Age Groups')
plt.show()
```

* Sucide rate between age group 35-54 are heighest.
* Sucide rate between age group 5-14 are lowest.

# 5. Which generation has highest number of suicide rate

```python
df["generation"].value_counts()
```

### * Average suicide numbers by generation

```python
df.groupby("generation", as_index = False)["suicides_no"].mean().sort_values(by="suicides_no",ascending = False)
```

### * Barplot for Suicide Numbers by Generations

```python
genDataSN = df.groupby("generation", as_index = False)["suicides_no"].mean().sort_values(by="suicides_no",ascending = False)
plt.figure(figsize=(8,5))
sns.barplot(x = "generation",
            y = "suicides_no",
            data = genDataSN)
plt.xlabel('Generation')
plt.ylabel('Number of Suicide')
plt.title('Suicide Numbers by Generations')
plt.show()
```

* We can clearly see that Boomers genration has heighest number of sucide rates.
